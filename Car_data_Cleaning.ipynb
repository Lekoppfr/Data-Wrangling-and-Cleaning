MSRP: The Manufacture's Suggested Retail price

Engine: the type of engine used

Passenger Doors: number of doors a car posseses

Base Curb Weight (lbs): Weigth of a car without passengers

Passenger Volume (ft³): people's sitting area capacity

Wheelbase (in): the distance between the rear and front tyre

Track Width: distance between two wheels on a same Axle

Front (in): the size front of a car

Height, Overall (in): the height of the car

Fuel Tank Capacity Approx (gal): the capity of fuel a tank can take


SAE Net Torque @ RPM: Revolutions per Minute Capacity as per Society of Automotive Engineers' standards

Fuel System: the type of fuel system set on a car

SAE Net Horsepower @ RPM: Revolutions per Minute capacity as per SOciety of Automotive Engineers' standards

Displacement: the measure of the total volume of air that is moved by the pistons in the engine’s cylinders

Trans Description Cont.: the description of a transmission

Trans Type: the type of transmission

Suspension Type - Front: the system of tires, tire air, springs, shock absorbers, and linkages that connects a vehicle to its wheels and allow relative movement between the two.

Suspension Type - Rear: the system of tires, tire air, springs, shock absorbers, and linkages that connects a vehicle to its wheels and allow relative movement between the two.

Air Bag-Frontal-Driver: the number of air bags infront of a car (driver)

Air Bag-Frontal-Passenger: the number of air bags infront of a car (passenger)

Air Bag-Passenger Switch (On/Off): number of Air Bag-Passenger Switch

Air Bag-Side Body-Front: the number of air bags infront of a car (body front)

Air Bag-Side Body-Rear:the number of air bags infront of a car (in the back)

Air Bag-Side Head-Front: the number of air bags infront of a car (infront side)

Air Bag-Side Head-Rear:the number of air bags infront of a car (in the back side)

Brakes-ABS: number of Short for Antilock Braking System (ABS) which is a braking technology that prevents tires from locking during emergency braking.

Child Safety Rear Door Locks: located on the edge of each rear door, this alows the rear door not to open if the inner door handle is pulled.

Daytime Running Lights:  these are low-wattage white lights automatically activate when you engage the vehicle’s ignition.

Traction Control:  is a safety system that detects when a car’s wheel slips and loses grip on the road. The system then automatically slows the slipping wheel’s speed.

Night Vision: is a system that uses an infrared sensor typically in the grille to look for warm objects in the roadway.

Rollover Protection Bars: is a safety feature. It consists of roll bars or roll cages that create a protective zone around the operator in the event of a rollover.

Fog Lamps: number of supplementary lights that can be used along with your headlights in low-visibility conditions

Parking Aid: a system that helps to park a car in  a garage

Tire Pressure Monitor: TPM uses pressure monitoring sensors within each tire that monitor specific pressure levels 

Back-Up Camera: devices that help drivers see behind their vehicles when reversing.

Stability Control: a stability control system monitors your steering wheel input and helps ensure the car goes where you want it to go.

Basic Miles/km: Car's mileage

Basic Years: the basic years of a car

Corrosion Miles/km: limit of corrosion miles for a car

Corrosion Years: period of years corrosion appears

Drivetrain Miles/km: the limit of miles a the drivetrain can take

Drivetrain Years: period of years a drivetrain can go

Turning Diameter - Curb to Curb (ft): It is the minimum space needed to turn the car around in a circular motion.

Front Wheel Material: front wheels are typically solid rubber, air-filled, or polyurethane. Each material offers a different level of durability, traction, and comfort.


Stabilizer Bar Diameter - Front (in): A stabilizer bar tries to keep the car's body flat by moving force from one side of the body to another.

Roadside Assistance Years: a warrant for road assistance

Roadside Assistance Miles/km: 

Manufacturer: the company that made the vehicle.

Model year: the year the model of the car was produced or launched

Category: the category that a car belongs to, electric or SUV etc..,

Front tire width: the size of the tire infront

Front tire aspect ratio: The aspect ratio is the height of your tire’s sidewall expressed as a percentage of its width.

Front tire speed ratings/cons.type:  It is the optimal speed that the tire can safely maintain over time.

Front tire rim size: rim diameter is the distance across the wheel. It’s the size commonly referred when talking about wheel size and it is crucial since it determines the size of tire needed and affects aspects like speedometer accuracy and vehicle handling.
